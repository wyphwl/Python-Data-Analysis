## 评估数据

# 评估整洁度

```python
original_data.sample(10)
```

# 评估数据干净度  

```python
original_data.info()
```

可见CustomerID和Description存在缺失值，InvoiceDate的数据类型应为日期，Customer数据类型应为字符串  

```python
original_data[original_data["Description"].isnull()]
```

```python
original_data[(original_data["Description"].isnull())&(original_data["UnitPrice"]!=0)]
```

结果发现产品名缺失的行单价全为0，即都为无效数据，应删除

```python
original_data[original_data["CustomerID"].isnull()]
```

 发现其他变量与其缺失相关性不大，且其本身不是必要数据，可保留

# 评估重复和不一致数据

```python
original_data["Country"].value_counts()
```

发现好多国家用不同名字，用describe对数值型数据进行快速统计了解

```python
original_data.describe()
```

```python
original_data[original_data["Quantity"]<0]
```

从结果来看，交易数量未负数的订单，发票编号都为C开头，可验证猜想

```python
original_data[(original_data["InvoiceNo"].str[0]!="C")&(original_data["Quantity"]<0)]
```

有数据，说明猜想是错误的，但单价还是为0

```python
original_data[(original_data["InvoiceNo"].str[0]!="C")&(original_data["Quantity"]<0)&(original_data["UnitPrice"]!=0)]
```

可知quantity为负数时，要么订单id开头为C,要么单价为0

```python
original_data[original_data["UnitPrice"]<0]
```

发现数据也无用，可删除

综上待办事项： 
    把InvoiceDate数据类型转换为日期时间  
    把Customer数据类型转换为字符串  
    把Description变量缺失的观察值删除  
    把Country变量值“USA”->"United States"  
    把Country变量值“UK”->"United Kingdom"  
    把Quantity值为负数的观察值删除  
    把UnitPrice值为负数的删除  

```python
cleaned_data=original_data.copy()
```

```python
cleaned_data["InvoiceDate"]=pd.to_datetime(cleaned_data["InvoiceDate"])
```

```python
cleaned_data["InvoiceDate"]
```

```python
cleaned_data["CustomerID"]=cleaned_data["CustomerID"].astype(str)
```

```python
cleaned_data["CustomerID"]=cleaned_data["CustomerID"].str.slice(0,-2)
```

```python
cleaned_data.dropna(subset=["Description"],inplace=True)
```

用来检验是否已经全部删除

```python
cleaned_data["Description"].isnull().sum()
```

```python
cleaned_data["Country"]=cleaned_data["Country"].replace({"USA":"United States"})
```

```python
len(cleaned_data[cleaned_data["Country"]=="USA"])
```

```python
cleaned_data=cleaned_data[cleaned_data["Quantity"]>0]
```

```python
cleaned_data=cleaned_data[cleaned_data["UnitPrice"]>0]
```

```python
cleaned_data.sample(50)
```

```python
cleaned_data.to_csv("data_cleaned.csv",index=False)
```
